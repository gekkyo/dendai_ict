```python
# 基本のライブラリを読み込む
import numpy as np
import pandas as pd
from scipy import stats

# グラフ描画
from matplotlib import pylab as plt
import seaborn as sns
%matplotlib inline
sns.set()

# グラフを横長にする
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

# 統計モデル
import statsmodels.api as sm
```

```python
# 普通にデータを読み込む
ts = pd.read_csv('sample.csv', sep = "\t", usecols = [0], dtype = float )
ts.head()

plt.plot(ts)
```

```python
# 自己相関を求める
ts_acf = sm.tsa.stattools.acf(ts, nlags=40)
ts_acf

#  自己相関のグラフ
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.tsa.plot_acf(ts, lags=30)
```

```python
# 偏自己相関を求める
df_pacf = sm.tsa.stattools.pacf(ts, nlags=20, method='ols')

# 偏自己相関を可視化する
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.tsa.plot_pacf(ts, lags=20)
```

```python
# 次数の推定
sm.tsa.arma_order_select_ic(ts, max_ar=20, max_ma=20, ic='aic')
```
